## 1. Normal Distribution

```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
from matplotlib import style
from IPython.core.display import HTML

%matplotlib inline
style.use("fivethirtyeight")
plt.rcParams["figure.figsize"] = (14, 7)
plt.figure(dpi=100)
```

```python
# PDF
x = np.linspace(-4,4,100)
y = stats.norm.pdf(np.linspace(-4,4,100)) / np.max(stats.norm.pdf(np.linspace(-3,3,100)))

plt.plot(x, y)
plt.fill_between(x, y, alpha=0.15)
plt.title("Standard Normal Distribution (PDF)")
plt.show()
```

```python
#CDF
y = stats.norm.cdf(np.linspace(-4,4,100))

plt.plot(x, y)
plt.title("Standard Normal Distribution (CDF)")
plt.show()
```
